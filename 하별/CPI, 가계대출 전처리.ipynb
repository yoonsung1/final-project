### CPI (소비자 물가 지수)

```python
import pandas as pd

# =========================
# 1. CPI 로드
# =========================
cpi_raw = pd.read_csv(
    'data/소비자물가지수(1996.1~).csv',
    encoding='cp949'
)

# =========================
# 2. 전국 평균만 사용 (중요)
# =========================
# '시도별' 컬럼에서 '전국' 행만 필터
cpi_raw = cpi_raw[cpi_raw['시도별'] == '전국']

# =========================
# 3. wide → long 변환
# =========================
cpi_long = cpi_raw.melt(
    id_vars='시도별',
    var_name='date',
    value_name='cpi'
)

# =========================
# 4. 날짜 파싱
# =========================
cpi_long['date'] = pd.to_datetime(cpi_long['date'], format='%Y.%m')

# =========================
# 5. 숫자형 변환
# =========================
cpi_long['cpi'] = (
    cpi_long['cpi']
    .astype(str)
    .str.replace(',', '')
    .astype(float)
)

# =========================
# 6. 기간 컷 (2004-01 ~)
# =========================
cpi = (
    cpi_long[cpi_long['date'] >= '2004-01-01']
    [['date', 'cpi']]
    .set_index('date')
    .sort_index()
)

# cpi.head()

cpi.tail()
```

### 가계대출

```python
print(hh_loan.columns)
```

```python
import pandas as pd

# 1. 데이터 로드
hh_raw = pd.read_csv(
    'data/가계대출_용도별.csv')

# 2. 총액 행만 선택 (예금취급기관 = 전체 가계대출)
hh_total = hh_raw[hh_raw['계정항목'] == '예금취급기관']

# 3. wide → long
hh_long = hh_total.melt(
    id_vars=['통계표', '계정항목', '단위', '변환'],
    var_name='date',
    value_name='hh_loan'
)

# 4. 날짜 파싱 (월별)
hh_long['date'] = pd.to_datetime(
    hh_long['date'],
    format='%Y/%m',
    errors='coerce'
)

# 5. 숫자 처리
hh_long['hh_loan'] = (
    hh_long['hh_loan']
    .astype(str)
    .str.replace(',', '', regex=False)
)

hh_long['hh_loan'] = pd.to_numeric(
    hh_long['hh_loan'],
    errors='coerce'
)

# 6. 기간 필터 (2008년 이후)
hh_loan = (
    hh_long
    .dropna(subset=['hh_loan'])
    .query("date >= '2008-01-01'")
    [['date', 'hh_loan']]
    .set_index('date')
    .sort_index()
)
```
